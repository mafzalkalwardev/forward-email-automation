```python
# ===================== IMPORTS =====================
import pandas as pd
import random
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import time
import os

# ===================== CONFIG =====================
CSV_FILE = 'carrier_details_interrupted.csv'  # your CSV file path
SENDER_EMAIL = 'ryanstills1.echovista@gmail.com'
SENDER_NAME = 'Apex Transit LLC'
PASSWORD = 'kpau npnd ehiw tyts'  # Your Gmail App Password
BATCH_DELAY = 10  # seconds between emails to reduce spam risk
LOG_FILE = 'sent_emails_log.csv'  # stores emails already sent
MAX_EMAILS = 100  # send only first 100 unsent emails

# ===================== SUBJECTS =====================
subjects = [
    "Big Earnings Alert: Premium Loads Waiting for You!",
    "Maximize Your Truck Income – Hot Loads Inside!",
    "🚛 Top-Paying Truck Loads – Don’t Miss Out!",
    "Unlock High-Paying Freight Opportunities Today!",
    "Your Truck Deserves These Profitable Loads!",
    "Instant Load Offers – Boost Your Revenue!",
    "Exclusive Freight Deals for Hardworking Truckers!",
    "High-Dollar Loads Ready for Immediate Pickup!",
    "Drive Your Profits Up with These Hot Loads!",
    "Load Up & Cash In: Premium Jobs Available!",
    "Don’t Miss These High-Paying Freight Runs!",
    "🚚 Profitable Loads You Can’t Ignore!",
    "Big Money Loads Waiting for Your Truck!",
    "Freight Goldmine: Top-Paying Routes Today!",
    "Load Offers That Fill Your Wallet Fast!",
    "Quick, High-Profit Loads – Only Today!",
    "Your Next Big Haul is Here – Start Earning!",
    "💰 Trucking Jobs with Maximum Returns!",
    "Premium Loads Just a Click Away!",
    "High-Earning Loads for Truckers Like You!"
]

# ===================== MESSAGES =====================
messages = [
    "Looking for dedicated carriers ready to earn big weekly! We have premium loads available for those willing to work 5+ days a week. Want a load today? Share your zip code.",
    "Serious carriers wanted! Earn top weekly gross with our premium loads. Work at least five days a week and start hauling today. Send your zip code to get started.",
    "Attention carriers! High-paying loads are available for drivers ready to commit 5 days a week. Share your zip code and get your load today.",
    "Premium loads alert! We’re looking for carriers who want to maximize their weekly earnings. Work 5 days a week and start today. Reply with your zip code.",
    "Carriers needed! Great weekly gross opportunities for those working five days a week. Premium loads available—share your zip code to get started today.",
    "Serious drivers only! Earn more with our premium load options. Work at least five days weekly. Send your zip code to receive your first load today.",
    "Looking for hardworking carriers! Premium loads available for those ready to work 5+ days a week. Want a load today? Reply with your zip code.",
    "Earn top weekly profits! We have premium loads for carriers working five days a week. Share your zip code to get your load today.",
    "Apex Transit LLC is looking for dedicated carriers! Premium load options available for 5-day/week drivers. Want to haul today? Send your zip code.",
    "Carriers, boost your weekly earnings! Premium loads available for those committed to working 5 days a week. Share your zip code to get started.",
    "Serious carriers wanted! Work five days a week and earn great weekly gross with our premium loads. Reply with your zip code to get a load today.",
    "High-paying loads available! Looking for carriers who can work at least 5 days a week. Want to pick up a load today? Share your zip code.",
    "Attention professional carriers! Premium loads are ready for drivers working 5+ days a week. Send your zip code and get a load today.",
    "Earn big weekly! We’re offering premium load options for carriers who work at least five days a week. Share your zip code to start hauling today.",
    "Dedicated carriers wanted! Premium loads with high weekly gross potential are available. Work 5 days a week and share your zip code to get a load today.",
    "Serious carriers, serious earnings! Premium loads ready for drivers working five days a week. Reply with your zip code for today’s load.",
    "Ready to maximize your weekly earnings? Premium loads available for committed carriers working 5+ days a week. Share your zip code to haul today.",
    "Looking for reliable carriers! Great weekly gross and premium loads available for 5-day/week drivers. Want a load today? Send your zip code.",
    "Carriers, make serious money! Premium loads ready for those willing to work five days a week. Share your zip code to get started today.",
    "High-profit loads for dedicated carriers! Work five days a week and earn top weekly gross. Reply with your zip code to pick up your load today."
]

# ===================== LOAD CSV =====================
df = pd.read_csv(CSV_FILE)

# ===================== RESUME SUPPORT =====================
if os.path.exists(LOG_FILE):
    sent_df = pd.read_csv(LOG_FILE)
    sent_emails = set(sent_df['Email'].tolist())
else:
    sent_emails = set()

# ===================== SEND EMAIL FUNCTION =====================
def send_email(to_email, to_name):
    subject = random.choice(subjects)
    message_template = random.choice(messages)
    message_body = f"Hi {to_name},\n\n{message_template}\n\nBest regards,\n{SENDER_NAME}"

    msg = MIMEMultipart()
    msg['From'] = f"{SENDER_NAME} <{SENDER_EMAIL}>"
    msg['To'] = to_email
    msg['Subject'] = subject
    msg.attach(MIMEText(message_body, 'plain'))

    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(SENDER_EMAIL, PASSWORD)
        server.send_message(msg)
        server.quit()
        print(f"✅ Email sent to {to_name} ({to_email}) with subject: {subject}")
        return True
    except Exception as e:
        print(f"❌ Failed to send email to {to_email}: {e}")
        return False

# ===================== MAIN LOOP =====================
sent_log = []
count = 0

for index, row in df.iterrows():
    email = row['Email']
    name = row['Legal Name']

    if email in sent_emails:
        print(f"⚠️ Already sent to {email}, skipping.")
        continue

    if send_email(email, name):
        sent_log.append(email)
        sent_emails.add(email)
        count += 1

    time.sleep(BATCH_DELAY)

    if count >= MAX_EMAILS:
        print(f"✅ Reached limit of {MAX_EMAILS} emails. Stopping.")
        break

# ===================== SAVE LOG =====================
if sent_log:
    log_df = pd.DataFrame(sent_log, columns=['Email'])
    if os.path.exists(LOG_FILE):
        existing_log = pd.read_csv(LOG_FILE)
        log_df = pd.concat([existing_log, log_df], ignore_index=True)
    log_df.to_csv(LOG_FILE, index=False)

print("✅ Batch sending complete and log updated.")
```
